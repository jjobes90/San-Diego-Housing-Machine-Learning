Column 2 - House construction year

```python
import pandas as pd
  
# defining function to correct year built values

def function(num):
  
    if num>0 and num<100:
        return (1900 + num)
    
    elif num>100 and num<1800:
        return "Unknown"
    
    elif num>1800 and num<2022:
        return num
    
    else:
        return "Unknown"

numeric_df['YrBlt']=numeric_df.YrBlt.apply(function)
```
